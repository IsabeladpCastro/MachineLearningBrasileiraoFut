```python

import pandas as pd
import matplotlib.pyplot as plt
import shap
import joblib
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.inspection import permutation_importance
import seaborn as sns
```

```python

df = pd.read_csv('brasileirao_featured.csv')
model = joblib.load('model_rf.pkl')

feature_cols = ['points', 'goals', 'goals_taken', 'played']
X = df[feature_cols]
y = df['place']
```

```python
explainer = shap.Explainer(model, X)
shap_values = explainer(X, check_additivity=False)

shap.summary_plot(shap_values.values, X)
plt.show()

```

```python

shap.plots.bar(shap_values)
plt.show()
```

```python

y_pred = model.predict(X)
cm = confusion_matrix(y, y_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.title("Matriz de ConfusÃ£o â€” Treino")
plt.show()
```

```python

print(classification_report(y, y_pred))
```

```python

importances = model.feature_importances_
importance_df = pd.DataFrame({'feature': feature_cols, 'importance': importances}).sort_values(by='importance', ascending=False)
importance_df
```

```python

perm = permutation_importance(model, X, y, n_repeats=10)
perm_df = pd.DataFrame({'feature': feature_cols, 'importance': perm.importances_mean}).sort_values(by='importance', ascending=False)
perm_df
```
