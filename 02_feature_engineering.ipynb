# Notebook 02 — Feature Engineering (Brasileirão Série A 2003–2024)

## Objetivo
Criar novas features esportivas úteis para prever desempenho.

```python
import pandas as pd
import numpy as np

DATA_CSV_PATH = 'brasileirao.csv'
df = pd.read_csv(DATA_CSV_PATH)
df.head()
```

## Feature: aproveitamento (%)

```python
df['aproveitamento'] = (df['points'] / df['played']) * 100
df.head()
```

## Feature: média de gols marcados

```python
df['media_gols_marcados'] = df['goals_taken'] / df['played']
df.head()
```

## Feature: média de gols sofridos

```python
df['media_gols_sofridos'] = df['goals_diff'] / df['played']
df.head()
```

## Salvar dataset processado

```python
df.to_csv('brasileirao_featured.csv', index=False)
```
